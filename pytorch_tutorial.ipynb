```python
import torch

# Data
x = torch.tensor([[1.0], [2.0], [3.0]])
y = torch.tensor([[3.0], [5.0], [7.0]])

# Model parameters
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

# Training loop
for i in range(100):
    # Forward
    y_pred = w * x + b

    # Loss (MSE)
    loss = ((y_pred - y)**2).mean()

    # Backward
    loss.backward()

    # Update weights
    with torch.no_grad():
        w -= 0.01 * w.grad
        b -= 0.01 * b.grad

    # Zero gradients
    w.grad.zero_()
    b.grad.zero_()

print(w, b)
```

```python
import torch.nn as nn
import torch.optim as optim

x = torch.tensor([[1.0], [2.0], [3.0]])
y = torch.tensor([[3.0], [5.0], [7.0]])

model = nn.Linear(1,1)

optimizer = optim.Adam(model.parameters() , lr = 0.01)

criterion = nn.MSELoss()

for i in range(100):
  y_pred = model(x)
  loss = criterion(y_pred , y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

print(list(model.parameters()))
```
