```python
import torch.nn as nn
```

```python
model = nn.Sequential(
    nn.Conv2d(1,8 , kernel_size = 3 , padding=2),
    nn.ReLU(),
    nn.MaxPool2d(2),

    nn.Conv2d(8,16 , kernel_size=3 , padding=2),
    nn.ReLU(),
    nn.MaxPool2d(2),

    nn.Flatten(),
    nn.Linear(16*7*7 ,10)
)
```
